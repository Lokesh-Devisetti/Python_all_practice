```python
# question : 2
# prove a given number is a perfect number:
# what is the perfect  number = 6 | the divisibles if we add should match the given number

a = 6
b = 0

for i in range(1,a):
  if a % i == 0:
    b = b + i

if a == b:
  print(f'{a} is a perfect number')
else:
  print(f'{a} is not a perfect number')
```

```python
# question : 3 | prove the given number is a prime number or not


a = 7
b = 0

for i in range(1,a+1): #1,2,3,4,5,6,7
  if a % i == 0:#7%1==0, 7%7=0
    b = b + 1 #0+1,1+1

if b == 2:
  print(f'{a} : is a Prime Number')
else:
  print(f'{a} : is not a prime number')
```

```python
# question :4 from value 1 to 100 find all the prime numbers:

c = []
for i in range(2,101):
  b = 0
  for j in range(1,i+1):
    if i % j == 0:
      b = b + 1
  if b == 2:
    c.append(i)
print(c)
```

```python
# Find the Given number is palindrome or not:

a = 888
b = a
count = 0 #8

while a > 0: #88>0,8>0
  res = a % 10 #888%10,88%10,8%10
  count = (count * 10) + res #(0*10)+8,(8*10)+8,(88*10)+8
  a = a // 10 #888//10,88//10,8//10

if b == count:
  print(f'{b} : is a palindrome number ')
else:
  print(f'{b} : is not a palindrome number ')
```

```python
# Prove a number is an Amstrong or not:

a = 9474
b = a
count = 0 #27,152,153

while a > 0: #153>0,15>0,1>0,0>0
  res = a % 10 #153%10,15%10,1%0
  count = count + (res * res * res) #0+(3*3*3),27+(5*5*5),152+(1*1*1)
  a = a // 10 #153//10,15//10,1//0

if b == count:
  print(f'{b} : is a Amstrong number ')
else:
  print(f'{b} : is not a Amstrong number ')
```

```python
# strong number: means the sum of factorial of each digit

a = 145
c = a
count = 0

while a > 0:
  res = a % 10
  b = 1
  for i in range(1,res+1):
    b = b * i
  count = count + b
  a = a // 10

if c == count:
  print(f'{c} : is a strong number')
else:
  print(f'{c} : is not a strong number')
```
