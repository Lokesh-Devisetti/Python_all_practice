```python
# know multiprocessing starts:
import multiprocessing
import os

def task_1():
  print('I belong to task_1')

def task_2():
  print('I belong to task_2')



if __name__ == '__main__':  # python is communicating with cpu
  print('The main CPU processor id :',os.getpid())

  child_1 = multiprocessing.Process(target=task_1,args=[])
  child_1.start()
  child_1.join()
```

```python
import time
import os

def fun(a):
  print(f'This function is running by : {os.getpid()}')
  for i in range(a):
    print(i)
    time.sleep(0.01)


if __name__ == '__main__':
  print(f'The Main Processor ID : {os.getpid()}')

  child_1 = multiprocessing.Process(target=fun, args=(10,))
  child_1.start() # start executing the sub processor
  child_1.join()  # main processor will wait till the work completed by child processor
  print(f'main processor ends : {os.getpid()}')
```

```python
# Do multiprocessing in python works together




import time
import os

def fun(a):
  print(f'This function is running by : {os.getpid()}')
  for i in range(a):
    print('c1',i)
    time.sleep(0.5)

def add(a):
  print(f'This function is running by : {os.getpid()}')
  for i in range(a):
    print('c2',i)
    time.sleep(0.5)



if __name__ == '__main__':
  print(f'The Main Processor ID : {os.getpid()}')

  child_1 = multiprocessing.Process(target=fun, args=(12,))
  child_2 = multiprocessing.Process(target=add, args=(12,))


  child_1.start() # start executing the child_1 processor
  child_2.start() # start executing the child_2 Processor


  child_1.join()  # main processor will wait till the work completed by child processor
  child_2.join()  # main processor will wait till the work completed by child processor

  print(f'main processor ends : {os.getpid()}')
```
