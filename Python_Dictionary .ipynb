
Dictionay is mutable -> Prove it

```python
# addition:

a = {'Name':'mani' , 'age':21 , 'salary':10}  # it will save in memory

a['location'] = 'bangalore' # this line will add a new key called location and its value is bangalore

print(a)
```

```python
 a={} # it is an empty dictionary

a['apple'] = 100
a['grapes'] = 200

print(a)
```

```python
# deletion :

a = {'Name':'loki' , 'age':21 , 'salary':10}

del a['age'] # i want to delete age column

print(a)
```

```python
# replace:

a = {'Name':'loki' , 'age':21 , 'salary':10}

a['age'] = 55

print(a)
```

**Dictionary built-in-functions**

```python
# clear : it will clear the entire data but varibale will be alive:

a = {'Name':'mani' , 'age':21 , 'salary':10}

a.clear() # it clear all the data

print(a)
```

```python
# copy: copy from one var to another var

a = {'Name':'mani' , 'age':21 , 'salary':10}

b = a.copy()

print(a)
print(b)
```

```python
# get : it is same like calling

a = {'Name':'mani' , 'age':21 , 'salary':10}

print(a['salary'])
```

```python
# get : it is same like calling

a = {'Name':'mani' , 'age':21 , 'salary':10}

print(a.get('salary'))
```

```python
# keys:print all the keys

a = {'Name':'A' , 'age':21 , 'salary':10}  # in this what are the keys:


print(a.keys())
```

```python
# values :print all the values

a = {'Name':'A' , 'age':21 , 'salary':10}  # in this what are the values:


print(a.values())
```

```python
# items: both keys and values


a = {'Name':'A' , 'age':21 , 'salary':10}


print(a.items())
```

```python
# addition:

a = {'Name':'A' , 'age':21 , 'salary':10}  # it will save in memory

a['location'] = 'bangalore' # this line will add a new key called location and its value is bangalore

print(a)
```

```python
# update: we can update to the existing varibale

a = {'Name':'A' , 'age':21 , 'salary':10}

a.update({"location":"Hyderabad"})

print(a)
```

```python
a = {} # it is an empty dictionary

a['apple'] = 100
a['grapes'] = 200

print(a)
```

```python
a = {}

a.update({"apple":100, "banana":200})

print(a)
```

```python
# pop and popitem:
# pop : mean which remove based on input:

a = {'Name':'A' , 'age':21 , 'salary':10}

a.pop('age') # it will remove age key and its value

print(a)
```

```python
# popitem : it will remove the last key and its value

a = {'Name':'A' , 'age':21 , 'salary':10}

a.popitem() # last key value will be removed

print(a)
```
