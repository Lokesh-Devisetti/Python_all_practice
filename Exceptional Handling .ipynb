- `Exceptional handling works with -> try block and except block `

- `syntax : `

      - Except:
          
          Access the Error here

```python
# know using exceptional Handling we can optimize the error:
# try -> the place where you will write the code -> if no error it is fine -> if error means it gives to expect block:


try:
  name = 'sharuk'
  print(NAME)


except Exception as e:
        print(e)
```

```python
try:
  a = 'sai'
  b = int(a)
  print(b)

except Exception as e:
  print(e)
```

```python
# finally -> there is a function in python which we can use while implementing Exceptional handling:

# finally block will execute if there is error and if there is no error also:
```

```python
try:
  a = 'sai'
  b = int(a)
  print(b)

except Exception as e:
  print(e)

finally:
  print("i will execute with the error also")
```

```python
# accept the exception directly:

try:
  a = 100
  b = 0
  print(a / b)


except ZeroDivisionError as e:
  print(f'The error is : {e}')

except ValueError as e:
  print(f'The error is : {e}')

finally:
  print(f'I will execute all the time')
```
