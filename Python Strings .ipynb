**Build_in_Functions in strings**

capitalize: this function will make starting letter from all the string as a capital
title: this function make each word starting letter as a capital
lower : is used to make capital to lower case [lower -> small letters]
upper : makes to capital letters - len : is used to find the number of characters in a string
replace : is used to replace old character with a new character
isupper : return boolean out come [True or False]
islower : returns boolean outcome [True or False]
isalpha : checks all alphabets or not
isnumeric : checks all numbers or not
isaplnum : checks numbers or alphates or not
Membership : return boolean outcomes
count : count the number of occurences
index : find the index of a character
