```python
def even(s,e):
  for i in range(s,e):
    if i % 2 == 0:
      print(i)

even(11,17)
even(21,27)
```

**Difference Between Local varibale and Global Varibale**

- `What is Global varibale : means we can use the varibale any where [the varibale which creates outside the funtion is know as Global varibale so that you can use it inside the function also ]`

- `Local varibale : the varibales which are defined inside the funciton only -> and they can use only inside the function we cant use them outside the  function:`

```python
d = 100  # this is a global variable

def fun():
  print(d + 100)

print(d)
fun()
```

```python
def fun():
  k = 100
  print(k) # local varibale


print(k)
fun()
```
