File Handling
File Handing means use to Handle the File using the concept modes
know we are going to work with .txt files -> use NotePad to create .txt Files
Modes IN FileHandling

w - is used to write the data into the file if the file is available -> if the file is not available w -> will create the file and will write the data into it:

r -> is used to read the data from the File -> if the File is not available it will through an error :

a -> is used to add the data into the File -> if the file is not available a -> will create the file and will add the data into it:

```python
# w  -> write the data into the file:

a = open('sharuk.txt' , 'w') # open the file or create the file

a.write("I am supporting India tomorrow for the final match")  # write the data into the file

a.close()  # you opened the File in 1st line so here after completing the operations we need to close it
```

```python
# empty file is there so please write data into it:

a = open('sachin.txt','w') # file is there so it will open the file

a.write("India is really working Hard") # it will write the data into the file

a.close()  # close the file
```

```python
# file the there and data also there then what happends:
# it is going to override the existing data

a = open('sachin.txt','w') # file is there so it will open the file

a.write("India is really working Hard") # in the file data is already available so the fresh data is going to override

a.close()  # close the file
```

```python
# r -> is used to read the data from the  file:

a = open('sachin.txt' , 'r') # checks the file if the file is available it reads the data or else it will through error

print(a.read())


a.close()
```

```python
# a -> append [add]

a = open('sachin.txt','a')  # open the file

a.write('I am from chennai') # write the data into file

a.close() # close the File
```

```python
# file is there but it is empty:

a = open('sachin.txt','a')  # open the file

a.write('I am from chennai') # write the data into file

a.close() # close the File
```

```python
# readline and readlines:
# readline -> is used to read only Few lines from the data
# readlines -> is used to read all the lines from the data in list Manner:


a = open('python.txt','r')

print(a.readline()) # readline will read only few lines of data

a.close()
```

```python
# readline and readlines:
# readline -> is used to read only Few lines from the data
# readlines -> is used to read all the lines from the data in list Manner:


a = open('python.txt','r')

print(a.readlines()) # readline will read only few lines of data

a.close()
```


r -> used to read the data

read -> it will read the entire data from the file exactly how it is in the file

readline : is used to read the data in few lines

readlines : is used to read the entire data in list manner
