```python
# know i will take the help of constructor and make the code more better


class checking:
  def __init__(self,x,y,a,b,c,d):
    self.x = x
    self.y = y
    self.a = a
    self.b = b
    self.c = c
    self.d = d

  def addition(self):
    return self.x + self.y

  def subtraction(self):
    return self.x - self.y

  def multiplication(self):
    return self.c * self.d

  def division(self):
    return self.x / self.y

  def common(self):
    print(self.addition()  , self.subtraction() , self.multiplication() , self.division())


kiran = checking(10,20,1,2,3,4)
kiran.common()
```

```python
# Inheritance -> means using other class functions or varibales:


class parent:
  def mumbai(self):
    print('I am the Parent and I live in mumbai')



class child(parent):
  def pakistan(self):
    print('I want to see Pakistan')


sai = child() # for the child class I created a object
sai.pakistan() # child class function
sai.mumbai() # parent class function
```

```python
# multilevel inheritance:

class grand_father:
  def tamilnadu(self):
    print('I used to stay in chennai')



class parent(grand_father):
  def mumbai(self):
    print('I am the Parent and I live in mumbai')



class child(parent):
  def pakistan(self):
    print('I want to see Pakistan')


sai = child() # for the child class I created a object
sai.pakistan() # child class function
sai.mumbai() # parent class function
sai.tamilnadu() # grand_father functions
```

```python
# multiple inheritance


class father:
  def tamilnadu(self):
    print('I used to stay in chennai')



class mother:
  def mumbai(self):
    print('I am the Parent and I live in mumbai')



class child(father,mother):
  def pakistan(self):
    print('I want to see Pakistan')


obj = child()
obj.pakistan()
obj.mumbai()
obj.tamilnadu()
```

```python
# Hierarical Inheritance


class parent:
  def tamilnadu(self):
    print('I used to stay in chennai')

class son1(parent):
  def mumbai(self):
    print('I am the Parent and I live in mumbai')

class son2(parent):
  def pakistan(self):
    print('I want to see Pakistan')

class son3(parent):
  def karnataka(self):
    print('Bangalore is best city')


obj = son3()
obj.karnataka()
obj.tamilnadu()
```

```python
# Hybrid Inheritance


class parent:
  def tamilnadu(self):
    print('I used to stay in chennai')

class son1(parent):
  def mumbai(self):
    print('I am the Parent and I live in mumbai')

class son2(parent):
  def pakistan(self):
    print('I want to see Pakistan')

class son3(son1,son2):
  def karnataka(self):
    print('Bangalore is best city')


obj = son3()
obj.karnataka()
obj.mumbai()
obj.pakistan()
obj.tamilnadu()
```
