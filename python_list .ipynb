**python_list_build_in_functions**

- `append : will add the value only at the end of the list `
- `copy : will copy data in a list to another varibale `
- `count : we can count the occurences `
- `index : used to find the index `
- `extend : we can add multiple values at the end of the list `
- `insert : we can insert the value based on the index `
-  `pop : will remove the last the value and also remove based on index `
- `remove : directly take input and remove them `
- `reverse : reverse a list same like [::-1]`
- `sort : ascending order or descending order `
- `clear : clear is used to clear the data but not deleting the varibale`.


```python
a=[12,23,45,56]
a.append(89)
print(a)
```

```python
a.extend([34,90,89])
print(a)
```

```python
a.insert(2,'manikanta')
print(a)
```

```python
del a[2]
print(a)
```

```python
a.pop()
print(a)
```

```python
a.pop(5)
print(a)
```
