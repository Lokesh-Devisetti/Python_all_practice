Rules to Follow when using set

Set is an unordered pair-> wee dont know how it will save in the memory.
Since sets are unordered -> it dont have index concept.
Sets wont accept duplicates.
sets accept every data type expect mutable data types -> since set is a mutable but it wont accept mutable inside it.

```python

a = {14,56,43,234,3435} # saves in the memory but we dont know how it saves
print(a) # call and print from the memory
print(type(a)) # type -> find out what variable is holding
```

```python
# set:  index is not possible in set concept:

a = {14,56,43,234,3435} # saves in the memory
print(a[1])
```

```python
# set wont accept duplicates:

a = {10,11,10,15,21,5,11,4}  # while saving in memory 10 and 11 will be only one time

print(a)
```

```python
# set wont accept mutable data types:

a = {10 , 12.55 , 'kiran' , (55,76)}

print(a)
```

```python
a = {10 , 12.55 , 'kiran' , [55,76]}

print(a)
```

**Sets are mutable**



*   Adding memory space for the existing variable
*   Deleting memory space for existing varibale



```python
# addition:

a = {10,21,55}

a[3] = 100

print(a)
```

Python sets not have index so, we can perform built-in-functions.

```python
# add : will take a value from the user and will add randomlly somewhere in the set:

a = {10,21,55} # know a is holding set:

a.add(100) # add function will add the value to the varible given

print(a)
```

```python
# update : it is like extend function in list -> they will take multiple values and add to the existing varibale

a = {10,21,55}

a.update([100,200,3000])  # here in the memory values are getting added :

print(a)
```

```python
# delete: using pop built in function we can remove a value

# pop


a = {10,21,55,100}

a.pop()  # randomlly it will remove one value:


print(a)
```

# remove and discard :

# discard : if the value is there it will remove or else it will keep calm and no error will be thrown
# remove : if the value is not there it will through an error

```python
#discard

a = {10,21,55,100}

a.discard(55) # want to remove value 55

print(a)
```

```python
#discard

a = {10,21,550,100}

a.discard(55) # want to remove value 55

print(a)
```

```python
# remove:

a = {10,21,55,100}

a.remove(55) # want to remove value 55

print(a)
```

```python
# remove:

a = {10,21,55,100}

a.remove(550) # want to remove value 55

print(a)
```



*   From above functions wecan say sets are mutable.



```python
# clear , copy , union , intersection :
```

```python
# clear: it will clear all the data inside the varibale but the varibale will be alive
```

```python
a = {10,21,55,100}

a.clear() # clear all the data in a but varibale will be alive

print(a)
```

```python
a = set()  # creating an empty set -> a is an empty set

a.add(10)

print(a)
```

```python
# copy: copy the data from one varibale to another varibale

a = {10,21,55,100}

b = a.copy()

print(a)
print(b)
```

```python
# union and intersection

a = {1,2,3,4,5,6}
b = {5,6,7,8}

print(a.union(b)) # a union b
```

```python
# union and intersection

a = {1,2,3,4,5,6}
b = {5,6,7,8}

print(a.intersection(b)) # a intersect b
```
