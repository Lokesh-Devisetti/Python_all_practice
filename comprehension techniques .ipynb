```python
for i in range(11,18):
  print(i)
```

```python
# List Comprehension:

print([i for i in range(11,18)])
```

```python
for i in range(501,521):
  if i % 2 == 0:
    print('even',i)
```

```python
# in list comprehension [for lop should be middle if should be right and print should be in starting]

print([i for i in range(501,521) if i % 2 == 0])
```

```python
for i in range(501,521):
  if i % 2 == 0:
    print('even',i)
  else:
    print('odd',i)
```

```python
# in list comprehension [for lop should be last if should be first and else should be middle ]
# the values which are even multiply by 10 odd value multiply by 100


print([i*10 if i % 2 == 0 else i*100 for i in range(11,18)])
```

```python
# Dictionary Comprehension:

a = {'apple':100 , 'banana':200}

for i in a:
  print(i)
```

```python
a = {'apple':100 , 'banana':200}
print(a['apple'])
```

```python
a = {'apple':100 , 'banana':200}

for i in a:
  print(a[i])
```

```python
a = {'apple':100 , 'banana':200}
print(a.items()) # items will give us both key and its value
```

```python
# with dict comprehension:

a = {'apple':100 , 'banana':200}

print({i:j for i,j in a.items()})
```

```python
a = {'apple':100 , 'banana':200}
# multiply apple with 100 and banana with 100


for i in a:
  a[i] = a[i] * 100


print(a)
```

```python
# with dict comprehension:

a = {'apple':100 , 'banana':200}

print({i:j*100 for i,j in a.items()})
```

```python
a = {'kiran':78 , 'vinay':32 , 'teja':67} # who are more than 40 make them pass

for i in a:
  if a[i] > 40:
    a[i] = 'pass'

print(a)
```

```python
a = {'kiran':78 , 'vinay':32 , 'teja':67}

print({i:'pass' for i,j in a.items() if j > 40})
```

```python
a = {'kiran':78 , 'vinay':32 , 'teja':67} # who are more than 40 make them pass

for i in a:
  if a[i] > 40:
    a[i] = 'pass'
  else:
    a[i] = 'fail'

print(a)
```

```python
a = {'kiran':78 , 'vinay':32 , 'teja':67}

print({i:('pass' if j > 40 else 'fail') for i,j in a.items()})
```
