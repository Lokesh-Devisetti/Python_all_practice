
Type Conversions

Converting int -> float and int -> str
Converting float -> int and Float -> str
Converting str -> int and str -> float

```python
a = 12 # save in memory
print(a) # calling and printing
print(type(a)) # checking variable type [what data type the variable is holding]
```

```python
# int -> float
# for type conversions use nick names only | int , str and float


a = 12
print(a)
print(type(a))
b = float(a)  # here in this line conversion is going on
print(b)
print(type(b))
```

```python
a = 465
b=40
str(a)
str(b)
```

```python
# int -> string
# for type conversions use nick names only | int , str and float

a = 12
print(a)
print(type(a))
b = str(a)  # here in this line conversion is going on
print(b)
print(type(b))
```

```python
# float -> int
# for type conversions use nick names only | int , str and float

a = 145.567
print(a)
print(type(a))
b = int(a)  # here in this line conversion is going on
print(b)
print(type(b))
```

```python
# float -> str
# for type conversions use nick names only | int , str and float

a = 24.5566
print(a)
print(type(a))
b = str(a)  # here in this line conversion is going on
print(b)
print(type(b))
```

```python
# string -> int
# for type conversions use nick names only | int , str and float

a = '45'
print(a)
print(type(a))
b = int(a)  # string can be converted to integer when all of them is an numbers only
print(b)
print(type(b))
```
